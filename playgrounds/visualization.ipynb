## Visualizations of the data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matchms.importing import load_from_mgf
```

```python
# Data
spec2vec = pd.read_csv('./source/embedding/tms_maccs/spec2vec.csv')
spec2vec.head(1)
```

```python
# Count of compounds for each InChIKey
plt.figure(figsize=(10, 5), dpi=600)

value_counts = spec2vec['inchi_key'].value_counts()
x = value_counts.index.str[:10]
y = value_counts.values
sns.barplot(x=x, y=y, palette='Blues_d', width=0.4)

plt.xlabel('InChi Key / Compound', fontsize=20, labelpad=10)
plt.ylabel('Count', fontsize=20)

plt.xticks([])

plt.savefig('./source/visulization/inchi_key_counts.pdf')
```

```python
spectrum = next(load_from_mgf('./source/dataset/Test dataset_TMS_RAW.mgf'))
spectrum
```

```python
spectrum.plot(figsize=(15, 5), dpi=600, grid=False)

plt.title('')
plt.xlabel('m/z', fontsize=18, labelpad=10)
plt.ylabel('Intensity', fontsize=18)
plt.grid(False)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplots_adjust(bottom=0.2)
plt.savefig('./source/visulization/spectrum.pdf')
```
