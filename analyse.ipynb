```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FINGERPRINT_FILE = './embeddings/tms_maccs_fingerprint.csv'
```

```python
df = pd.read_csv(FINGERPRINT_FILE)
df.drop(columns=['name', 'inchi'], inplace=True)
df.set_index('inchikey', inplace=True)
```

```python
# plot number of 1s per column
fig, ax = plt.subplots(figsize=(24, 8))
sns.barplot(x=df.columns, y=df.sum(), ax=ax, palette="crest")

ax.set_title('Number of 1s per column')
ax.set_xlabel('MACCS fingerprint bit')
ax.set_ylabel('Number of 1s')

plt.xticks(rotation=90, fontsize=8)
plt.show()
```

```python
# plot number of 1s per row with horizontal barplot
fig, ax = plt.subplots(figsize=(6, 12))

sns.barplot(x=df.sum(axis=1), y=df.index, ax=ax, palette="crest")

ax.set_title('Number of 1s per row')

plt.yticks(fontsize=8)
plt.show()
```
