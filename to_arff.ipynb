```python
import arff
import pandas as pd
```

## Fingerprint

```python
FILE = "./embeddings/tms_maccs_fingerprint.csv"
OUTPUT = f"{'.'.join(FILE.split('.')[:-1])}.arff"
```

```python
df = pd.read_csv(FILE)
```

```python
df.head()
```

```python
df.dropna(axis=0, how='any', inplace=True)
```

```python
df.rename(columns={"Name": "name", "InChI": "inchi", "InChI Key": "inchikey"}, inplace=True)
```

```python
df["name"] = df["name"].str.strip()
df["inchi"] = df["inchi"].str.strip()
df["inchikey"] = df["inchikey"].str.strip()
```

```python
df.tail(20)
```

```python
df.to_csv(FILE, index=False)
```

```python
arff.dump(OUTPUT, df.values, relation='tms_maccs_fingerprint', names=df.columns)
```

### Merged

```python
FILE = "./embeddings/tms_maccs_fingerprint_and_embedding.csv"
OUTPUT = f"{'.'.join(FILE.split('.')[:-1])}.arff"
```

```python
df = pd.read_csv(FILE)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum().sum()
```

```python
bool_fingerprints = df.filter(regex="^fingerprint_").astype(bool)
df[bool_fingerprints.columns] = bool_fingerprints
```

```python
df.info()
```

```python
arff.dump(OUTPUT, df.values, relation='tms_maccs_fingerprint_and_embedding', names=df.columns)
```
