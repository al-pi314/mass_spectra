```python
import arff
import pandas as pd
```

## Fingerprint

```python
FILE = "./embeddings/tms_maccs_fingerprint.csv"
OUTPUT = f"{'.'.join(FILE.split('.')[:-1])}.arff"
```

```python
df = pd.read_csv(FILE)
```

```python
df.head()
```

```python
df.iloc[:, 3:] = df.iloc[:, 3:].astype(bool)
```

```python
df.rename(columns={"Name": "name", "InChI": "inchi", "InChI Key": "inchikey"}, inplace=True)
```

```python
df["name"] = df["name"].str.strip()
df["inchi"] = df["inchi"].str.strip()
df["inchikey"] = df["inchikey"].str.strip()
```

```python
df.head()
```

```python
df.to_csv(FILE, index=False)
```

```python
arff.dump(OUTPUT, df.values, relation='tms_maccs_fingerprint', names=df.columns)
```
